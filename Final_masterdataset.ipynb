```python
import pandas as pd
import os
import matplotlib.pyplot as plt
#from sklearn.impute import SimpleImputer
```

```python
os.listdir('data')
```

```python
temp = pd.read_pickle('./data/temp.pkl')
sun = pd.read_pickle('./data/sun.pkl')
wind = pd.read_pickle('./data/wind.pkl')
electricity_bal = pd.read_pickle('./data/electricitybalance.pkl')
prices = pd.read_pickle('./data/price_test_set.pkl')
offshore = pd.read_pickle('./data/offshore_test.pkl')
```

```python
temp.columns = ['temp']
```

```python
md = pd.read_pickle('./data/master_dataset_24112019.pkl')
```

```python
data = pd.concat([prices.resample('H').mean(),
                  electricity_bal.resample('H').mean(),
                  sun.resample('H').mean(),
                  temp.resample('H').mean(),
                  wind.resample('H').mean()], axis=1, join='inner')
```

```python
data['hour'] = data.index.hour
data['weekday'] = data.index.weekday
```

```python
data = pd.concat([data, offshore.resample('H').mean()], axis=1, join='inner')
```
