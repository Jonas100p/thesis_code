## t+24

```python
df24 = pd.DataFrame()
```

```python
data['dk2_lag'] = data['dk2'].shift()
data = data.dropna(how='any', axis=0)
```

```python
data_train = data[data.index.date < datetime.date(2015, 1, 1)]
data_test = data[data.index.date > datetime.date(2017, 1, 1)]
```

```python

for i in range(24):
    data_train_24 = data_train[data_train.index.hour == i]
    endog24 = data_train_24['dk2']
    exog24 = data_train_24.drop(['dk2'], axis=1)
    
    data_test_24 = data_test[data_test.index.hour == i]
    data_train_24 = data_train_24[['dk2', '-1', '-2', '-3', '-4', '-5', '-6', '-7', '-8']]
    mod24 = ARMA(endog=endog24, exog=exog24, order=(1,0))
    endog24 = data_train_24['dk2']
    exog24 = data_train_24.drop(['dk2'], axis=1)
    res24 = mod24.fit()
    
    data_test_24['preds'] = res24.params[0] \
                        + res24.params[1]*data_test_24['dk2_lag'] + \
                        + res24.params[2]*data_test_24['-1'] \
                        + res24.params[3]*data_test_24['-2'] \
                        + res24.params[4]*data_test_24['-3'] \
                        + res24.params[5]*data_test_24['-4'] \
                        + res24.params[6]*data_test_24['-5'] \
                        + res24.params[7]*data_test_24['-6'] \
                        + res24.params[8]*data_test_24['-7'] \
                        + res24.params[9]*data_test_24['-8']
    
    df24 = pd.concat([df24, data_test_24], join='outer', axis=0)

df24 = df24.sort_index()
```
