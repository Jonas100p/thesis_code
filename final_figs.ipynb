```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lstm1 = pd.read_pickle('lstm_no.pkl')
lstm24 = pd.read_pickle('lstm_24no.pkl')
lstm48 = pd.read_pickle('lstm_48no.pkl')
lstm1_pred = pd.read_pickle('lstm_pred.pkl')
lstm24_pred = pd.read_pickle('lstm_24pred.pkl')
lstm48_pred = pd.read_pickle('lstm_48pred.pkl')
combo1 = pd.read_pickle('combo1.pkl')
combo24 = pd.read_pickle('combo24.pkl')
combo48 = pd.read_pickle('combo48.pkl')
arma1 = pd.read_pickle('arma1.pkl')
arma24 = pd.read_pickle('arma24.pkl')
arma48 = pd.read_pickle('arma48.pkl')
rf1 = pd.read_pickle('rfdf1.pkl')
rf24 = pd.read_pickle('rfdf24.pkl')
rf48 = pd.read_pickle('rfdf48.pkl')
last_period1 = pd.read_pickle('last_period_1.pkl')
last_period24 = pd.read_pickle('last_period_24.pkl')
last_period48 = pd.read_pickle('last_period_48.pkl')
```

```python
plt.figure(figsize=(24, 16))
plt.title('Prediction error on percentile (t+1)', fontsize=20)
plt.xlabel('percentile')
plt.ylabel('mean absolute error DKK/MWh')
plt.plot(last_period1, label='last_period', color='grey', linestyle='dashed')
plt.plot(lstm1, label='lstm', color='orange')
plt.plot(lstm1_pred, label='lstm + pred', color='blue')
plt.plot(arma1, label='AR', color='red')
plt.plot(rf1, label='rf', color='green')
plt.plot(combo1, label='rf+lstm', color='black')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(24, 16))
plt.title('Prediction error on percentile (t+24)', fontsize=20)
plt.xlabel('percentile')
plt.ylabel('mean absolute error DKK/MWh')
plt.plot(last_period24, label='last_period', color='grey', linestyle='dashed')
plt.plot(lstm24,  label='lstm', color='orange')
plt.plot(lstm24_pred, label='lstm + pred', color='blue')
plt.plot(arma24, label='AR', color='red')
plt.plot(rf24, label='rf', color='green')
plt.plot(combo24, label='rf+lstm', color='black')
plt.legend()
plt.show()
```
